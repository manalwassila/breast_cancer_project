```python
import os
import shutil

# chemin du dataset téléchargé
dataset_path = "C:/Users/manal/Downloads/archive"

# dossiers destination
benign_path = "../data/train/benign"
malignant_path = "../data/train/malignant"

os.makedirs(benign_path, exist_ok=True)
os.makedirs(malignant_path, exist_ok=True)

# parcourir dataset
for root, dirs, files in os.walk(dataset_path):
    for file in files:
        if file.endswith(".png"):
            
            source = os.path.join(root, file)
            
            if "class0" in file:
                shutil.copy(source, benign_path)
                
            elif "class1" in file:
                shutil.copy(source, malignant_path)

print("✅ Tri terminé")
```
